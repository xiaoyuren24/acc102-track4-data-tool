```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import f_oneway, linregress
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', 50)
# Load the unemployment data
print("Loading unemployment data...")
df = pd.read_csv('unemployment.csv', encoding='utf-8')
print(f"✅ Data loaded: {df.shape[0]} rows, {df.shape[1]} columns")
# Display the first few rows
print("📋 First 5 rows of raw data:")
display(df.head())
# Transform wide format to long format
print("Transforming data from wide to long format...")

# Identify year columns
year_columns = [col for col in df.columns if '[YR' in str(col)]
id_columns = ['Series Name', 'Series Code', 'Country Name', 'Country Code']

# Convert to long format
df_long = pd.melt(
    df,
    id_vars=id_columns,
    value_vars=year_columns,
    var_name='Year_Column',
    value_name='Unemployment_Rate'
)

# Extract year from Year_Column
df_long['Year'] = df_long['Year_Column'].str.extract(r'(\d{4})').astype(int)

# Handle missing values
df_long['Unemployment_Rate'] = pd.to_numeric(df_long['Unemployment_Rate'], errors='coerce')

# Rename columns
df_long = df_long.rename(columns={
    'Series Name': 'Indicator_Name',
    'Series Code': 'Indicator_Code',
    'Country Name': 'Country_Name',
    'Country Code': 'Country_Code'
})

# Select required columns
df_long = df_long[['Indicator_Name', 'Indicator_Code', 'Country_Name', 'Country_Code', 'Year', 'Unemployment_Rate']]

print(f"✅ Data transformed: {df_long.shape[0]} rows, {df_long.shape[1]} columns")
print(f"Time period: {df_long['Year'].min()} to {df_long['Year'].max()}")
print(f"Countries: {df_long['Country_Name'].nunique()}")
# Display transformed data
print("📋 Sample of transformed data:")
display(df_long.head(10))
# Basic information about the data
print("📊 DATA OVERVIEW")
print("=" * 60)
print(f"Total Records: {len(df_long):,}")
print(f"Number of Countries: {df_long['Country_Code'].nunique()}")
print(f"Year Range: {df_long['Year'].min()} - {df_long['Year'].max()}")
print(f"Unique Indicators: {len(df_long['Indicator_Name'].unique())}")

# Data quality
total_records = len(df_long)
missing_records = df_long['Unemployment_Rate'].isnull().sum()
valid_records = total_records - missing_records
print(f"\n📈 DATA QUALITY:")
print(f"Valid Records: {valid_records:,} ({valid_records/total_records*100:.1f}%)")
print(f"Missing Records: {missing_records:,} ({missing_records/total_records*100:.1f}%)")
# Countries information
print("🌍 COUNTRIES COVERED:")
print("=" * 50)
countries = df_long[['Country_Name', 'Country_Code']].drop_duplicates().sort_values('Country_Name')
for _, row in countries.iterrows():
    print(f"  {row['Country_Code']}: {row['Country_Name']}")
    # Statistical summary
print("📊 STATISTICAL SUMMARY (Unemployment Rate %):")
print("=" * 60)
valid_rates = df_long['Unemployment_Rate'].dropna()
print(f"Mean: {valid_rates.mean():.2f}%")
print(f"Median: {valid_rates.median():.2f}%")
print(f"Standard Deviation: {valid_rates.std():.2f}%")
print(f"Minimum: {valid_rates.min():.2f}%")
print(f"Maximum: {valid_rates.max():.2f}%")
print(f"25th Percentile: {valid_rates.quantile(0.25):.2f}%")
print(f"75th Percentile: {valid_rates.quantile(0.75):.2f}%")
# Country-wise statistics
print("📊 COUNTRY-WISE STATISTICS:")
print("=" * 100)
print(f"{'Country':25} {'Years':>12} {'Avg Rate':>12} {'Min':>10} {'Max':>10} {'Std Dev':>10}")
print("-" * 100)

for country_code in df_long['Country_Code'].unique():
    country_data = df_long[df_long['Country_Code'] == country_code]
    valid_data = country_data.dropna(subset=['Unemployment_Rate'])
    country_name = country_data['Country_Name'].iloc[0]
    print(f"{country_name:25} {len(valid_data):>12} {valid_data['Unemployment_Rate'].mean():>11.2f}% "
          f"{valid_data['Unemployment_Rate'].min():>9.2f}% {valid_data['Unemployment_Rate'].max():>9.2f}% "
          f"{valid_data['Unemployment_Rate'].std():>9.2f}%")
# Create pivot table for easier analysis
unemp_pivot = df_long.pivot_table(
    index='Year',
    columns='Country_Name',
    values='Unemployment_Rate',
    aggfunc='mean'
)

print("📊 Unemployment Rate Data Overview (%):")
display(unemp_pivot.round(2).tail(10))
# Plot unemployment trends for all countries
plt.figure(figsize=(14, 8))

# Use a color palette
colors = plt.cm.tab10(np.linspace(0, 1, len(unemp_pivot.columns)))

for i, country in enumerate(unemp_pivot.columns):
    plt.plot(unemp_pivot.index, unemp_pivot[country], 
            label=country, linewidth=2.5, color=colors[i])

plt.title('Unemployment Rate Trends by Country (1990-2025)', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Unemployment Rate (%)', fontsize=12)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
plt.close()
# Interactive plot using Plotly
fig = go.Figure()

# Create a color palette for Plotly
plotly_colors = px.colors.qualitative.Set3

for i, country in enumerate(unemp_pivot.columns):
    color_idx = i % len(plotly_colors)
    fig.add_trace(go.Scatter(
        x=unemp_pivot.index,
        y=unemp_pivot[country],
        mode='lines',
        name=country,
        line=dict(width=3, color=plotly_colors[color_idx]),
        hovertemplate=f"<b>{country}</b><br>" +
                       "Year: %{x}<br>" +
                       "Unemployment Rate: %{y:.2f}%<br>" +
                       "<extra></extra>"
    ))

fig.update_layout(
    title='Interactive Unemployment Rate Trends (1990-2025)',
    xaxis_title='Year',
    yaxis_title='Unemployment Rate (%)',
    hovermode='x unified',
    height=600,
    template='plotly_white',
    legend=dict(
        yanchor="top",
        y=0.99,
        xanchor="left",
        x=1.05
    )
)

fig.show()
# Get data for specific years of interest
years_of_interest = [1990, 2000, 2010, 2020, 2025]

for year in years_of_interest:
    year_data = df_long[df_long['Year'] == year].dropna(subset=['Unemployment_Rate'])
    if not year_data.empty:
        year_data = year_data.sort_values('Unemployment_Rate', ascending=False)
        
        print(f"\n📊 Unemployment Rate Comparison - {year}:")
        print("=" * 70)
        print(f"{'Country':25} {'Rate (%)':>15}")
        print("-" * 70)
        for _, row in year_data.iterrows():
            print(f"{row['Country_Name']:25} {row['Unemployment_Rate']:14.2f}%")
            # Bar chart for unemployment rate comparison
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for idx, year in enumerate([1990, 2010, 2025]):
    year_data = df_long[df_long['Year'] == year].dropna(subset=['Unemployment_Rate'])
    year_data = year_data.sort_values('Unemployment_Rate', ascending=True)
    
    # Color coding
    colors = []
    for rate in year_data['Unemployment_Rate']:
        if rate < 5:
            colors.append('#2ecc71')  # Green
        elif rate < 8:
            colors.append('#f39c12')  # Orange
        elif rate < 12:
            colors.append('#e67e22')  # Dark Orange
        else:
            colors.append('#e74c3c')  # Red
    
    bars = axes[idx].barh(year_data['Country_Name'], year_data['Unemployment_Rate'], 
                        color=colors, alpha=0.8)
    
    axes[idx].set_title(f'Unemployment Rates - {year}', fontsize=14, fontweight='bold')
    axes[idx].set_xlabel('Unemployment Rate (%)', fontsize=11)
    
    # Add value labels
    for bar in bars:
        width = bar.get_width()
        axes[idx].text(width + 0.1, bar.get_y() + bar.get_height()/2,
                    f'{width:.1f}%', ha='left', va='center', fontsize=9)

plt.suptitle('Unemployment Rate Comparison Across Years', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
plt.close()
```

```python
# Create processed dataset with additional metrics
df_processed = df_long.copy()

# Calculate annual changes
df_processed['Annual_Change'] = df_processed.groupby('Country_Code')['Unemployment_Rate'].pct_change() * 100

# Calculate 5-year moving average
df_processed['5_Year_MA'] = df_processed.groupby('Country_Code')['Unemployment_Rate'].transform(
    lambda x: x.rolling(window=5, min_periods=1).mean()
)

# Calculate unemployment level
def get_unemployment_level(rate):
    if pd.isna(rate):
        return 'Unknown'
    elif rate < 3:
        return 'Very Low'
    elif rate < 5:
        return 'Low'
    elif rate < 8:
        return 'Medium'
    elif rate < 12:
        return 'High'
    else:
        return 'Very High'

df_processed['Unemployment_Level'] = df_processed['Unemployment_Rate'].apply(get_unemployment_level)

# Save to CSV
output_filename = 'unemployment_comprehensive_analysis.csv'
df_processed.to_csv(output_filename, index=False, encoding='utf-8-sig')

print(f"✅ Processed data saved to: {output_filename}")
print(f"Processed data shape: {df_processed.shape}")
print(f"\n📋 Sample of processed data:")
display(df_processed[['Country_Name', 'Year', 'Unemployment_Rate', 
                     'Annual_Change', '5_Year_MA', 'Unemployment_Level']].head(10))
```
